#### 安装依赖

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### 导入数据

```python
khexcel_file = 'result/in_YC_A_H.xlsx'
khdf = pd.read_excel(khexcel_file)
kzexcel_file = 'result/in_YC_A_Z.xlsx'
kzdf = pd.read_excel(kzexcel_file)
kqexcel_file = 'result/in_YC_A_Q.xlsx'
kqdf = pd.read_excel(kqexcel_file)

mhexcel_file = 'result/in_YC_A_H-mlp.xlsx'
mhdf = pd.read_excel(mhexcel_file)
mzexcel_file = 'result/in_YC_A_Z-mlp.xlsx'
mzdf = pd.read_excel(mzexcel_file)
mqexcel_file = 'result/in_YC_A_Q-mlp.xlsx'
mqdf = pd.read_excel(mqexcel_file)
```

```python
htime = np.arange(0 ,119)
khin_True_A = khdf.iloc[1:, 7].values
khin_Predicted_A = khdf.iloc[1:, 8].values
mhin_True_A = mhdf.iloc[1:, 7].values
mhin_Predicted_A = mhdf.iloc[1:, 8].values


ztime = np.arange(0 ,99)
kzin_True_A = kzdf.iloc[1:, 7].values
kzin_Predicted_A = kzdf.iloc[1:, 8].values
mzin_True_A = mzdf.iloc[1:, 7].values
mzin_Predicted_A = mzdf.iloc[1:, 8].values



qtime = np.arange(0 ,79)
kqin_True_A = kqdf.iloc[1:, 7].values
kqin_Predicted_A = kqdf.iloc[1:, 8].values
mqin_True_A = mqdf.iloc[1:, 7].values
mqin_Predicted_A = mqdf.iloc[1:, 8].values
```

#### 绘图

```python

# #color=(32/255, 94/255, 167/255)  ###蓝色
# #color=(232/255, 68/255, 69/255)  ###红色
# #color=(246/255, 99/255, 28/255)  ###橙色
# #color=(1/255, 138/255, 103/255)  ###绿色
```

```python
last = 80
timeend = 119
plt.figure(figsize=(12, 8))

plt.axvspan(0, 80, color='gray', alpha=0.1)
plt.axvspan(80, 120, color='khaki', alpha=0.3)
plt.text(50, 0.008, 'Train', fontsize=12, ha='center')
plt.text(100, 0.008, 'Prediction', fontsize=12, ha='center')

plt.plot(htime[:last], khin_True_A[:last], 'o', 
         color=(57/255, 127/255, 199/255), markerfacecolor='none', markersize=6, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255), label='Training Data') 
plt.plot(htime[last:timeend], khin_True_A[last:timeend], '*', 
         color=(57/255, 127/255, 199/255), markerfacecolor='none', markersize=8, markeredgewidth=1,
         markeredgecolor=(57/255, 127/255, 199/255), label='Prediction Data') 
plt.plot(htime[:last+1], khin_True_A[:last+1], '-', color=(232/255, 68/255, 69/255), linewidth=2, label='RMM')
plt.plot(htime[:last+1], khin_Predicted_A[:last+1], '-', color=(1/255, 138/255, 103/255), linewidth=2, label='Learned(KAN-Ude)') 
plt.plot(htime[:last+1], mhin_Predicted_A[:last+1], '-', color='black', linewidth=2, label='Learned(MLP)') 

plt.plot(htime[last:timeend], khin_True_A[last:timeend], '--', color=(232/255, 68/255, 69/255), linewidth=2, label='RMM')
plt.plot(htime[last:timeend], khin_Predicted_A[last:timeend], '--', color=(1/255, 138/255, 103/255), linewidth=2, label='Learned(KAN-Ude)') 
plt.plot(htime[last:timeend], mhin_Predicted_A[last:timeend], '--', color='black', linewidth=2, label='Learned(MLP)') 

plt.xlabel('Time',fontsize=18)
plt.ylabel('Daily New Infection', fontsize=18)
# plt.title(r"Fraction: $\frac{I}{1+I}$ MLP&KAN",fontsize=20)
# plt.title(r'Fuction: $\beta$S=0.4 MLP&KAN (H)',fontsize=20)
plt.xlim(-0.5, 120)
plt.legend(loc='upper left')
plt.savefig('in_YC_A_H.tif')

# plt.plot(htime[last:timeend], khin_True_A[last:timeend], 'o', color=(246/255, 99/255, 28/255), markersize=8, markeredgecolor='black', label='True data') 
```

```python
last = 60
timeend = 120
plt.figure(figsize=(12, 4))

plt.axvspan(0, 60, color='gray', alpha=0.1)
plt.axvspan(60, 120, color='khaki', alpha=0.3)
plt.text(30, 0.012, 'Train', fontsize=12, ha='center')
plt.text(90, 0.012, 'Prediction', fontsize=12, ha='center')

plt.plot(ztime[:last], kzin_True_A[:last], 'o', 
         color=(57/255, 127/255, 199/255),markerfacecolor='none', markersize=6, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255), label='True data') 
plt.plot(ztime[:last+1], kzin_True_A[:last+1], '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Mechanism model')
plt.plot(ztime[:last+1], kzin_Predicted_A[:last+1], '-', color=(1/255, 138/255, 103/255), linewidth=2, label='Network KAN') 
plt.plot(ztime[:last+1], mzin_Predicted_A[:last+1], '-', color='black', linewidth=2, label='Network MLP') 


plt.plot(htime[last:timeend], kzin_True_A[last:timeend], 'o', 
         color=(57/255, 127/255, 199/255),markerfacecolor='none', markersize=6, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255), label='True data') 
plt.plot(htime[last:timeend], kzin_True_A[last:timeend], '--', color=(232/255, 68/255, 69/255), linewidth=2, label='Mechanism model')
plt.plot(htime[last:timeend], kzin_Predicted_A[last:timeend], '--', color=(1/255, 138/255, 103/255), linewidth=2, label='Network KAN') 
plt.plot(htime[last:timeend], mzin_Predicted_A[last:timeend], '--', color='black', linewidth=2, label='Network MLP') 

plt.xlabel('Time',fontsize=18)
plt.ylabel('Incidence', fontsize=18)
# plt.title(r"Fraction: $\frac{I}{1+I}$ MLP&KAN",fontsize=20)
plt.title(r'Fuction: $\beta$S=0.4 MLP&KAN (Z)',fontsize=20)
plt.xlim(-0.5, 120)
plt.legend(loc='upper left')
plt.savefig('in_YC_A_Z.tif')
```
