```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
k32excel_file = 'result/KAN/kan_t.xlsx'
m32excel_file = 'result/mlp/mlp_t.xlsx'
k16excel_file = 'result/KAN/kan_fenshi.xlsx'
m16excel_file = 'result/mlp/mlp_fenshi.xlsx'
k8excel_file = 'result/KAN/kan_E.xlsx'
m8excel_file = 'result/mlp/mlp_E.xlsx'


time = np.arange(0 ,119)

k8df = pd.read_excel(k8excel_file)
m8df = pd.read_excel(m8excel_file)
k8_True_A = k8df.iloc[1:, 7].values
k8_Predicted_A = k8df.iloc[1:, 8].values
m8_True_A = m8df.iloc[1:, 7].values
m8_Predicted_A = m8df.iloc[1:, 8].values

k16df = pd.read_excel(k16excel_file)
m16df = pd.read_excel(m16excel_file)
k16_True_A = k16df.iloc[1:, 7].values
k16_Predicted_A = k16df.iloc[1:, 8].values
m16_True_A = m16df.iloc[1:, 7].values
m16_Predicted_A = m16df.iloc[1:, 8].values

k32df = pd.read_excel(k32excel_file)
m32df = pd.read_excel(m32excel_file)
k32_True_A = k32df.iloc[1:, 7].values
k32_Predicted_A = k32df.iloc[1:, 8].values
m32_True_A = m32df.iloc[1:, 7].values
m32_Predicted_A = m32df.iloc[1:, 8].values
```

```python
fig, axs = plt.subplots(3, 2, figsize=(16, 12))

# plt.subplot(4, 2, 1)
# plt.axvspan(0, 120, color='gray', alpha=0.1)
# plt.plot(time, m5_True_A, 'o', 
#          markerfacecolor='none', markersize=4, markeredgewidth=2,
#          markeredgecolor=(57/255, 127/255, 199/255), 
#          label='Data') 
# plt.plot(time, m5_Predicted_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Learned Curve') 
# plt.xlabel('Time(days)', fontsize=18)
# plt.ylabel('Daily New Infection', fontsize=18)
# plt.title(r'Mass action $\beta$SI(MLP)', fontsize=16)
# plt.legend()
# plt.xlim(-0.5, 120)

# plt.subplot(4, 2, 2)
# plt.axvspan(0, 120, color='gray', alpha=0.1)
# plt.plot(time, k5_True_A, 'o', 
#          markerfacecolor='none', markersize=4, markeredgewidth=2,
#          markeredgecolor=(57/255, 127/255, 199/255),
#          label='Data') 
# plt.plot(time, k5_Predicted_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Learned Curve') 
# plt.xlabel('Time(days)', fontsize=18)
# plt.ylabel('Daily New Infection', fontsize=18)
# plt.title(r'Mass action $\beta$SI(KAN)', fontsize=16)
# plt.legend()
# plt.xlim(-0.5, 120)


plt.subplot(3, 2, 1)
plt.axvspan(0, 120, color='gray', alpha=0.1)
plt.plot(time, m8_True_A, 'o', 
         markerfacecolor='none', markersize=4, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255), 
         label='Data') 
plt.plot(time, m8_Predicted_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Learned Curve') 
plt.xlabel('Time(days)', fontsize=18)
plt.ylabel('Daily New Infection', fontsize=18)
plt.title(r'Incidence rate with media effects $\beta$$e^{- \alpha I}$SI(KAN)',fontsize=16)
plt.legend()
plt.xlim(-0.5, 120)

plt.subplot(3, 2, 2)
plt.axvspan(0, 120, color='gray', alpha=0.1)
plt.plot(time, k8_True_A, 'o', 
         markerfacecolor='none', markersize=4, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255),
         label='Data') 
plt.plot(time, k8_Predicted_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Learned Curve') 
plt.xlabel('Time(days)', fontsize=18)
plt.ylabel('Daily New Infection', fontsize=18)
plt.title(r'Incidence rate with media effects $\beta$$e^{- \alpha I}$SI(MLP)',fontsize=16)
plt.legend()
plt.xlim(-0.5, 120)


plt.subplot(3, 2, 3)
plt.axvspan(0, 120, color='gray', alpha=0.1)
plt.plot(time, m16_True_A, 'o', 
         markerfacecolor='none', markersize=4, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255), 
         label='Data') 
plt.plot(time, m16_Predicted_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Learned Curve') 
plt.xlabel('Time(days)', fontsize=18)
plt.ylabel('Daily New Infection', fontsize=18)
plt.title(r'Nonlinear saturated incidence rate $\beta \frac{S I^l}{1 + \alpha I^h}$(MLP)',fontsize=16)
plt.legend()
plt.xlim(-0.5, 120)

plt.subplot(3, 2, 4)
plt.axvspan(0, 120, color='gray', alpha=0.1)
plt.plot(time, k16_True_A, 'o', 
         markerfacecolor='none', markersize=4, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255),
         label='Data') 
plt.plot(time, k16_Predicted_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Learned Curve') 
plt.xlabel('Time(days)', fontsize=18)
plt.ylabel('Daily New Infection', fontsize=18)
plt.title(r'Nonlinear saturated incidence rate $\beta \frac{S I^l}{1 + \alpha I^h}$(KAN)',fontsize=16)
plt.legend()
plt.xlim(-0.5, 120)


plt.subplot(3, 2, 5)
plt.axvspan(0, 120, color='gray', alpha=0.1)
plt.plot(time, m32_True_A, 'o', 
         markerfacecolor='none', markersize=4, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255), 
         label='Data') 
plt.plot(time, m32_Predicted_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Learned Curve') 
plt.xlabel('Time(days)', fontsize=18)
plt.ylabel('Daily New Infection', fontsize=18)
plt.title(r'Incidence rate with control effects $\beta\left(\left(c_{0}-c_{b}\right) e^{-r t}+c_{b}\right)$SI(MLP)',fontsize=16)
plt.legend()
plt.xlim(-0.5, 120)

plt.subplot(3, 2, 6)
plt.axvspan(0, 120, color='gray', alpha=0.1)
plt.plot(time, k32_True_A, 'o', 
         markerfacecolor='none', markersize=4, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255),
         label='Data') 
plt.plot(time, k32_Predicted_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Learned Curve') 
plt.xlabel('Time(days)', fontsize=18)
plt.ylabel('Daily New Infection', fontsize=18)
plt.title(r'Incidence rate with control effects $\beta\left(\left(c_{0}-c_{b}\right) e^{-r t}+c_{b}\right)$SI(KAN)',fontsize=16)
plt.legend()
plt.xlim(-0.5, 120)


labels = ['(A)', '(B)', '(C)', '(D)', '(E)', '(F)']
# 为每个子图添加标签
for i, ax in enumerate(axs.flat):
    ax.text(-0.1, 1.1, labels[i], transform=ax.transAxes, 
            fontsize=16, fontweight='normal', va='top', ha='right')

plt.tight_layout()

plt.savefig('KANVSMLP.tif')

plt.show()

# 获取当前的Axes对象
# ax = plt.gca()
# # 关闭上边框和右边框
# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)
```
