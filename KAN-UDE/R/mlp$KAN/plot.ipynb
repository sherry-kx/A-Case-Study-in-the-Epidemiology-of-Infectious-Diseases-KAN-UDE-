```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
kexcel_file = 'result/KAN/kan2-128-1.xlsx'
mexcel_file = 'result/mlp/mlp2-128-1.xlsx'
kdf = pd.read_excel(kexcel_file)
mdf = pd.read_excel(mexcel_file)
time = np.arange(0 ,119)
k_True_A = kdf.iloc[1:, 7].values
k_Predicted_A = kdf.iloc[1:, 8].values
m_True_A = mdf.iloc[1:, 7].values
m_Predicted_A = mdf.iloc[1:, 8].values
```

```python
plt.figure(figsize=(16, 4))

plt.subplot(1, 2, 1)
plt.axvspan(0, 120, color='gray', alpha=0.1)
# plt.axvspan(40, 60, color='khaki', alpha=0.3)
plt.text(60, 0.005, 'Train', fontsize=16, ha='center')
# plt.text(50, 0.8, 'Inference', fontsize=12, ha='center')

plt.plot(time, m_True_A, 'o', 
         markerfacecolor='none', markersize=4, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255), 
         label='True data') 
plt.plot(time, m_Predicted_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Network data') 
plt.xlabel('Time', fontsize=18)
plt.ylabel('Incidence', fontsize=18)
plt.title('MLP-2-128-1', fontsize=20)
plt.legend()
plt.xlim(-0.5, 120)

ax = plt.gca()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.subplot(1, 2, 2)
plt.axvspan(0, 120, color='gray', alpha=0.1)
# plt.axvspan(40, 60, color='khaki', alpha=0.3)
plt.text(60, 0.005, 'Train', fontsize=16, ha='center')
# plt.text(50, 0.12, 'Inference', fontsize=12, ha='center')

plt.plot(time, k_True_A, 'o', 
         markerfacecolor='none', markersize=4, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255),
         label='True data') 
plt.plot(time, k_Predicted_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Network data') 
plt.xlabel('Time', fontsize=18)
plt.ylabel('Incidence', fontsize=18)
plt.title('KAN-2-128-1', fontsize=20)
plt.legend()
plt.xlim(-0.5, 120)

ax = plt.gca()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.tight_layout()

plt.savefig('figure/2-128-1.tif')

plt.show()
```
