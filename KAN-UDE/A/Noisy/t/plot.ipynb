```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
xexcel_file = 'ZS-t.xlsx'
xdf = pd.read_excel(xexcel_file)
xzexcel_file = 'ZS-zt.xlsx'
xzdf = pd.read_excel(xzexcel_file)

yexcel_file = 'ZS-ti.xlsx'
ydf = pd.read_excel(yexcel_file)
yzexcel_file = 'ZS-zti.xlsx'
yzdf = pd.read_excel(yzexcel_file)
yzsexcel_file = 'ZS-ti-z.xlsx'
yzsdf = pd.read_excel(yzsexcel_file)

```

```python
xztime = np.arange(0 ,79)
xTrue_A = xdf.iloc[1:, 7].values  #
xEquation_A = xdf.iloc[1:, 7].values #
xzZ_True_A = xzdf.iloc[1:, 7].values  #
xPredicted_A = xdf.iloc[1:, 8].values  #
xzPredicted_A = xzdf.iloc[1:, 8].values  #


ytime = np.arange(0 ,79)
yTrue_A = ydf.iloc[1:, 7].values  
yEquation_A = ydf.iloc[1:, 7].values 
yZ_True_A = yzsdf.iloc[1:, 7].values  
yPredicted_A = ydf.iloc[1:, 8].values  
yzPredicted_A = yzdf.iloc[1:, 4].values  

```

```python
plt.figure(figsize=(12, 4))

plt.axvspan(0, 80, color='gray', alpha=0.1)
plt.text(40, 0.015, 'Train', fontsize=12, ha='center')

plt.plot(xztime, xTrue_A, 'o', 
         color=(57/255, 127/255, 199/255),markerfacecolor='none', markersize=6, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255), label='True data') 
plt.plot(xztime, xEquation_A, '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Mechanism model')
plt.plot(xztime, xzZ_True_A, '.', color='black', markersize=8, label='Noise data') 
plt.plot(xztime, xPredicted_A, '-', color=(1/255, 138/255, 103/255), linewidth=2, label='Network model') 
plt.plot(xztime, xzPredicted_A, '-', color=(246/255, 99/255, 28/255), linewidth=2, label='Noise Network model') 
plt.xlabel('Time',fontsize=18)
plt.ylabel('Incidence', fontsize=18)
# plt.title(r"Fraction: $\frac{I}{1+I}$ MLP&KAN",fontsize=20)
plt.title(r'Poisson noise Train Fuction: $5.0+10e^{(-0.2t)}$',fontsize=20)
plt.xlim(-0.5, 80)
# plt.legend(loc='upper left')
plt.legend()
plt.savefig('xunlian.tif')
```

```python
last = 30
timeend = 80
plt.figure(figsize=(12, 4))

plt.axvspan(0, 30, color='gray', alpha=0.1)
plt.axvspan(30, 80, color='khaki', alpha=0.3)
plt.text(15, 0.02, 'Train', fontsize=12, ha='center')
plt.text(60, 0.02, 'Prediction', fontsize=12, ha='center')

plt.plot(ytime[:last], yTrue_A[:last], 'o', 
         color=(57/255, 127/255, 199/255),markerfacecolor='none', markersize=6, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255), label='True data') 
plt.plot(ytime[:last+1], yEquation_A[:last+1], '-', color=(232/255, 68/255, 69/255), linewidth=2, label='Mechanism model')
plt.plot(ytime[:last+1], yPredicted_A[:last+1], '-', color=(1/255, 138/255, 103/255), linewidth=2, label='Network model') 
plt.plot(ytime[:last+1], yzPredicted_A[:last+1], '-', color=(246/255, 99/255, 28/255), linewidth=2, label='Noise Network model')
plt.plot(ytime[:last-1], yZ_True_A, '.', color='black', markersize=8, label='Noise data') 


plt.plot(ytime[last:timeend], yTrue_A[last:timeend], 'o', 
         color=(57/255, 127/255, 199/255),markerfacecolor='none', markersize=6, markeredgewidth=2,
         markeredgecolor=(57/255, 127/255, 199/255), label='True data') 
plt.plot(ytime[last:timeend], yEquation_A[last:timeend], '--', color=(232/255, 68/255, 69/255), linewidth=2, label='Mechanism model')
plt.plot(ytime[last:timeend], yPredicted_A[last:timeend], '--', color=(1/255, 138/255, 103/255), linewidth=2, label='Network model') 
plt.plot(ytime[last:timeend], yzPredicted_A[last:timeend], '--', color=(246/255, 99/255, 28/255), linewidth=2, label='Noise Network model') 

plt.xlabel('Time',fontsize=18)
plt.ylabel('Incidence', fontsize=18)
# plt.title(r"Fraction: $\frac{I}{1+I}$ MLP&KAN",fontsize=20)
plt.title(r'Poisson noise Prediction Fuction: $5.0+10e^{(-0.2t)}$',fontsize=20)
plt.xlim(-0.5, 80)
plt.legend()
plt.savefig('YC.tif')

# plt.plot(htime[last:timeend], khin_True_A[last:timeend], 'o', color=(246/255, 99/255, 28/255), markersize=8, markeredgecolor='black', label='True data') 
```
